```python
import numpy as np #NOTE: Since we already imported from sympy, we will use a  import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
X1, X2 = np.meshgrid(np.arange(-3, 3.01, .25), np.arange(-3, 3.01, .25)) # adju X1p = 1*X1+1*X2
X1p = 2*X1-X2
X2p = 3*X1-2*X2
#Normalize the arrows by dividing by their magnitude (focus on direction) U=1/(X1p**2+X2p**2)**(0.5)*X1p
U=1/(X1p**2+X2p**2)**(0.5)*X1p
V=1/(X1p**2+X2p**2)**(0.5)*X2p
plt.figure()
plt.title('Phase Plane for [[2,-1],[3,-2]]')
Q = plt.quiver(X1, X2, U, V) # draws the normalized arrows at (X,Y) with slope dYdX
```
